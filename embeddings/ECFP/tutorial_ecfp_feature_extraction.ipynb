```python
smiles
```

Get unique SMILES 

```python
unique_SMILES = np.unique(smiles.SMILES)
unique_SMILES
```

Create a molecule object for each SMILE

```python
mols = [Chem.MolFromSmiles(mol) for mol in unique_SMILES]
```

Extract Morgan Fingerprints

```python
df = np.stack([np.array(AllChem.GetMorganFingerprintAsBitVect(m,2,nBits=1024)) for m in mols])
```

Save Morgan Fingerprints to dataframe

```python
df = pd.DataFrame(df)
df.index = smiles.SMILES
df.to_csv('./data/embeddings.csv')
```
